# Call xy2sn function ot translate the data points to stream normal coordinates

sn, cl_out = xy2sn(x, y, cl_x, cl_y, rMax, **kwargs**)

Transforms XY coordinates to a new curvilinear coord. systems based on an input stream centerline.

Parameters
----------
**x** : 1-D array of X coordinates of the data points (can also be a column form an existing dataframe)\
**y** : 1-D array of y coordinates of the data points (can also be a column form an existing dataframe)

**cl_x** : 1-D array of X coordinates of the centerline (can also be a column form an existing dataframe)\
**cl_y** : 1-D array of Y coordinates of the centerline (can also be a column form an existing dataframe)

**r_max** : a search radius away from the centerline to search for points
- single float value - uniform search distance for all segments
- 1-D array - allows for varing search distances downstream
- 2-D array - allows for varing search distances downstream on both the right and left bank n X 2 (col1 = rt bank, col2 = left bank)

**nDiscr** : number of segments to resample the centerline to *[default = 1 meter downstream]*\
*See Notes below*
* this determines the accuracy of the outputs - more is better...to point
* enter 0 for default upsampling @ 2x the downstream distance
    * *example* - input CL length = 25m > nDiscr == 50

**kwargs**: keyword arguments = additional inputs

*to use add to the call xy2sn(x, y, cl_x, cl_y, r, keyword1 = value, keyword2 = value...)*

* nFilt : number of filtering iterations *(default = 5)*
* order : polynomial order of the filter *(default = 3)*
* window : number of points to include in the filter window *(default = 5)*
* plots : True or False *(default True)* - provides some simple graphs during the processing

Returns
-------
**snOut** : dataframe of the same length as the input XY data (columns = DS, XS)\
**clOut** : dataframe containing the smoothed and resampled centerline. Used for sn2xy. Save this to a CSV file!!!


```python
# basic inputs, uniform rMax (1), 0 for default nDiscr - no other options
sn, clOut = strNorm.xy2sn(data_pts.X,data_pts.Y,centerLine.X,centerLine.Y,1,0)
```

## Variable Search Distances

The "inside" bends in the centerline cause the search radius polygons to be compressed.

This can cause the search radii to overlap, the code *is not* smart enough to fix this and the last polygon a point intersects with will be its output values for `ds` and `xs` 

To avoid the search polygons overlapping you can input variable search distances into `xy2sn`as an array\
I suggest doing this in a GIS as you prepare your data - or by trial and error with the code.
* the length needs to match the length of the centerline array
* a 1-d array will vary the search radii symmetrically down the length of the centerline
* a 2-d array will vary the search radii independantly for each "bank" [left and right looking downstream]
    * the first column will be the left bank
    * the second column will be the right bank

```python
# load CL with variible search radii
cl_path = 'D:/Dropbox/Python/streamNormal/test_data/cl_variable.csv'
centerLine = pd.read_csv(cl_path)

# search radii array, transpose to make columns (code will check for transposed and fix)
searchR = np.array([centerLine.LR,centerLine.RR]).T

# variable search (both banks), 400 for nDiscr
sn, clOut = strNorm.xy2sn(data_pts.X,data_pts.Y,centerLine.X,centerLine.Y,searchR,400, verbose=False)
```

## Notes

`clOut.to_csv("path")`

#### nDiscr - number of segments to resample the centerline
* This variable is critical for proper calculations...
    * nDiscr should be > number of nodes in the input centerline
    * Too few centerline points can cause transformation errors (see example below)
    * A good starting point is about 10x the number of nodes on the input centerline

### Example of the effects of different nDiscr values on outputs

```python
# number of nodes in the input centerline
nodes = centerLine.shape[0]
print("number of input CL nodes: %i"%nodes)

# ** For the example data the data points are mathematically offset from the centerline
#    (a sin function) by 0.5 units. The output XS values are more accurate the closer to
#    +/- 0.5

# 1x input nodes
sn_1x, cl_1x = strNorm.xy2sn(data_pts.X,data_pts.Y,centerLine.X,centerLine.Y,1,nodes,plots=False, verbose=False)

# 2x input nodes
sn_2x, cl_2x = strNorm.xy2sn(data_pts.X,data_pts.Y,centerLine.X,centerLine.Y,1,nodes*2,plots=False, verbose=False)

# 3x input nodes
sn_3x, cl_3x = strNorm.xy2sn(data_pts.X,data_pts.Y,centerLine.X,centerLine.Y,1,nodes*3,plots=False, verbose=False)

# 4x input nodes
sn_4x, cl_4x = strNorm.xy2sn(data_pts.X,data_pts.Y,centerLine.X,centerLine.Y,1,nodes*4,plots=False, verbose=False)
```
